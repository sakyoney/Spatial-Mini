```python
import scanpy as sc

adata = sc.read_visium(
    path=str(base),
    count_file="filtered_feature_bc_matrix.h5",
    load_images=False
)

print("n_obs,n_vars:", adata.n_obs, adata.n_vars)
print("obs cols (first 20):", adata.obs.columns.tolist()[:20])
print("obsm keys:", list(adata.obsm.keys()))
print("uns keys:", sorted(adata.uns.keys()))
```

```python
import pandas as pd
import numpy as np

pos_path = base/"spatial/tissue_positions.csv"

df = pd.read_csv(pos_path)
if "barcode" not in df.columns:
    df = pd.read_csv(
        pos_path,
        header=None,
        names=["barcode","in_tissue","array_row","array_col","pxl_row_in_fullres","pxl_col_in_fullres"]
    )
df = df.set_index("barcode")

common = adata.obs_names.intersection(df.index)
print("common barcodes:", len(common), "/", adata.n_obs)

adata = adata[common].copy()
adata.obs = adata.obs.join(df.loc[common])

adata.obsm["spatial"] = adata.obs[["pxl_row_in_fullres","pxl_col_in_fullres"]].to_numpy().astype(np.float32)

print("obsm keys now:", list(adata.obsm.keys()))
print("spatial shape:", adata.obsm["spatial"].shape)
print(adata.obs[["in_tissue","array_row","array_col","pxl_row_in_fullres","pxl_col_in_fullres"]].head())
```

```python
adata_t = adata[adata.obs["in_tissue"] == 1].copy()
print("in_tissue spots:", adata_t.n_obs, "/", adata.n_obs)
```

```python
adata_t.var_names_make_unique()

import scanpy as sc
sc.pp.calculate_qc_metrics(adata_t, inplace=True)

print(adata_t.obs[["total_counts","n_genes_by_counts"]].describe().loc[["min","50%","max"]])
```

```python
sc.pp.normalize_total(adata_t, target_sum=1e4)
sc.pp.log1p(adata_t)

sc.pp.highly_variable_genes(adata_t, flavor="seurat_v3", n_top_genes=3000)
adata_t = adata_t[:, adata_t.var["highly_variable"]].copy()
print("HVG matrix:", adata_t.shape)
```

```python
sc.pp.scale(adata_t, max_value=10)
sc.tl.pca(adata_t, svd_solver="arpack")

sc.pp.neighbors(adata_t, n_neighbors=15, n_pcs=30)
sc.tl.umap(adata_t)
```
